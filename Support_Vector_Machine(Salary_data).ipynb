```python
import pandas as pd
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
train = pd.read_csv('/content/drive/MyDrive/datasets/SalaryData_Train(1).csv')
train
```

```python
test = pd.read_csv('/content/drive/MyDrive/datasets/SalaryData_Test(1).csv')
test
```

```python
test.info()
```

```python
train.columns
```

```python
col_cat = ['workclass', 'education', 'educationno', 'maritalstatus',
           'occupation', 'relationship', 'race', 'sex', 'native', 'Salary']
```

```python
for i in col_cat:
    print(i, train[i].value_counts())
```

```python
train1 = train.copy()
test1 = test.copy()
```

```python
train1 = train1.drop('educationno', axis = 1)
test1 = test1.drop('educationno', axis = 1)
```

```python
train1.columns
```

```python
test1.columns
```

```python
train1 = train1.apply(LabelEncoder().fit_transform)
train1.head()
```

```python
test1 = test1.apply(LabelEncoder().fit_transform)
test1.head()
```

```python
train1.shape
```

```python
x_train = train1.iloc[:,:-1]
y_train = train1.iloc[:,-1]
print(x_train.head())
y_train.head()
```

```python
x_test = test1.iloc[:,:-1]
y_test = test1.iloc[:,-1]
print(x_test.head())
y_test.head()
```

### Support Vector Machine Model Building

```python
model = SVC()
model.fit(x_train , y_train)
```

```python
preds = model.predict(x_test)
acc = accuracy_score(y_test, preds) * 100
```

```python
print('accuracy is', acc)
confusion_matrix(y_test, preds)
```

```python
model1 = SVC(kernel = 'linear', gamma = 5, C = 0.1)
model1.fit(x_train , y_train)
preds1 = model1.predict(x_test)
acc1 = accuracy_score(y_test, preds1) * 100
print('accuracy is', acc1)
confusion_matrix(y_test, preds1)
```

```python
model2 = SVC(kernel = 'rbf', gamma = 10, C = 0.8)
model2.fit(x_train , y_train)
preds2 = model2.predict(x_test)
acc2 = accuracy_score(y_test, preds2) * 100
print('accuracy is', acc2)
confusion_matrix(y_test, preds2)
```
